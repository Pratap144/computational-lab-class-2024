```python
# Import necessary libraries
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt

# Load the MD trajectory of beta amyloid from the Data folder
traj = md.load('Data/1hz3_T310.stepid25000000.every100ps.nowater.xtc', top='Data/1hz3_T310.start.nowater.gro')

# Part (b): Compute end-to-end distance and radius of gyration at each time

# Select the backbone atoms for calculating the end-to-end distance
backbone_indices = traj.topology.select("backbone")
end_to_end_distances = md.compute_distances(traj, [[backbone_indices[0], backbone_indices[-1]]])[:, 0]

# Compute the radius of gyration
radii_of_gyration = md.compute_rg(traj)

# Part (c): Plot the end-to-end distance and radius of gyration vs. time

# Convert time to nanoseconds (assuming the timestep is in picoseconds)
time = traj.time / 1000  # Convert from ps to ns

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(time, end_to_end_distances, label='End-to-End Distance (nm)', color='blue')
plt.plot(time, radii_of_gyration, label='Radius of Gyration (nm)', color='green')
plt.xlabel('Time (ns)')
plt.ylabel('Distance (nm)')
plt.title('End-to-End Distance and Radius of Gyration vs. Time')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save the plot to a file
plt.savefig('end_to_end_and_rg_vs_time.png')

# Show the plot
plt.show()

# Part (d): Plot normalized histograms of end-to-end distance and radius of gyration

# Plot histogram for end-to-end distance
plt.figure(figsize=(10, 6))
plt.hist(end_to_end_distances, bins=50, density=True, alpha=0.7, color='blue', label='End-to-End Distance')
plt.xlabel('End-to-End Distance (nm)')
plt.ylabel('Probability Density')
plt.title('Normalized Histogram of End-to-End Distance')
plt.grid(True)
plt.tight_layout()

# Save the histogram to a file
plt.savefig('end_to_end_histogram.png')

# Show the histogram
plt.show()

# Plot histogram for radius of gyration
plt.figure(figsize=(10, 6))
plt.hist(radii_of_gyration, bins=50, density=True, alpha=0.7, color='green', label='Radius of Gyration')
plt.xlabel('Radius of Gyration (nm)')
plt.ylabel('Probability Density')
plt.title('Normalized Histogram of Radius of Gyration')
plt.grid(True)
plt.tight_layout()

# Save the histogram to a file
plt.savefig('rg_histogram.png')

# Show the histogram
plt.show()
```
